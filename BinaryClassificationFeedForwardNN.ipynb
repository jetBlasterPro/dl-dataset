```python
import pandas as pd
```

```python
df= pd.read_csv("pima-indians-diabetes.csv")
```

```python
df.head(3)
```

```python
df.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifiernn=Sequential()
```

```python
classifiernn.add(Dense(units=10,activation="relu",input_dim=8))
classifiernn.add(Dense(units=8,activation="relu"))
classifiernn.add(Dense(units=1,activation="sigmoid"))
```

```python
classifiernn.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X.head(3)
```

```python
y.head(1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.25, random_state=1)
```

```python
xtrain.shape
```

```python
classifiernn.fit(xtrain,ytrain, epochs=500, batch_size=20) #576/20
```

```python
predictions= classifiernn.predict(xtest)
```

```python
predictions.shape
```

```python
predictions
```

```python
class_label=[]
for i in range(predictions.shape[0]):
    if(predictions[i]>0.5):
        class_label.append(1)
    else:
        class_label.append(0)
```

```python
class_label
```

```python
from sklearn.metrics import accuracy_score
```

```python
accu = accuracy_score(ytest,class_label)
```
